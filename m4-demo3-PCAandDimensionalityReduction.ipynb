```python
import pandas as pd
```

```python
#download link: https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv

wine_data = pd.read_csv('winequality-white.csv',
                       names=['Fixed Activity',
                           'Volatile Acidity',
                           'Citric Acid',
                           'Residual Sugar',
                           'Chlorides',
                           'Free Sulfur Dioxide',
                           'Total Sulfur Dioxide',
                           'Density',
                           'pH',
                           'Sulphates',
                           'Alcohol',
                           'Quality'   
                       ],
                       skiprows=1,
                       sep=r'\s*;\s*', engine='python')
wine_data.head()
```

```python
wine_data['Quality'].unique()
```

```python
X = wine_data.drop('Quality', axis=1)
Y = wine_data['Quality']

from sklearn import preprocessing
X = preprocessing.scale(X)

from sklearn.model_selection import train_test_split

X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.svm import LinearSVC

clf_svc = LinearSVC(penalty='l1', dual=False, tol=1e-3)
clf_svc.fit(X_train, Y_train)
```

```python
accuracy = clf_svc.score(x_test, y_test)
print(accuracy)
```

```python
# A heatmap allows us to see what features are strongly
# correlated with one another

import matplotlib.pyplot as plt
import seaborn as sns

corrmat = wine_data.corr()
f, ax = plt.subplots(figsize=(7,7))
sns.set(font_scale=0.9)
sns.heatmap(corrmat, vmax=.8, square=True, annot=True,
           fmt='.2f', cmap="winter")
plt.show()
```

```python
from sklearn.decomposition import PCA

#reduced n_components value to X from 11 -> 9 -> 6 -> 1
pca = PCA(n_components=1, whiten=True)
X_reduced = pca.fit_transform(X)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('Dimension')
plt.ylabel('Explain Variance Ratio')
plt.show()
```

```python
X_train, x_test, Y_train, y_test = train_test_split(X_reduced, 
                                                   Y,test_size=0.2,random_state=0)
clf_svc_pca = LinearSVC(penalty='l1', dual=False, tol=1e-3)
clf_svc_pca.fit(X_train, Y_train)
```

```python
accuracy = clf_svc_pca.score(x_test, y_test)
print(accuracy)
```
